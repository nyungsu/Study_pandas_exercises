### Step 1. Import the necessary libraries

```python
import pandas as pd
from collections import Counter
import plotly.express as px
# import matplotlib.pyplot as plt


# # set this so the graphs open internally
# %matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url = f'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
df = pd.read_csv(url,sep='\t')
df.shape
```

### Step 4. See the first 10 entries

```python
df.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
Counter(df['item_name']).most_common(5)
common_list = [item[0] for item in Counter(df['item_name']).most_common(5)]

temp = df[df['item_name'].isin(common_list)]
temp
fig = px.bar(temp,x='item_name',)
fig.update_layout(title='Most ordered items',
                  xaxis=dict(title='item name'),
                  yaxis=dict(title='number of order'))
fig.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

$ 1

```python
def changing_str2float(x:str):
    x = list(x)
    x = x[1:]
    x = ''.join(x)
    return float(x)

df['item_price'] = df['item_price'].apply(changing_str2float)
```

$ 2

```python
new_item_price = [float(ele[1:]) for ele in df.item_price]
```

```python
df.groupby('order_id').sum()

fig = px.scatter(df.groupby('order_id').sum(),
                 x='quantity',
                 y='item_price')
fig.update_layout(title='Number of items ordered per order price',
                  xaxis=dict(title='items ordered'),
                  yaxis=dict(title='order price'))
fig.show()
```
